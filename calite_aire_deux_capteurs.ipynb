```python
# Librairies
import folium  #Cercles colorés
import sys
import numpy
import pandas as pd #Excel
import numpy as np
from folium.plugins import MiniMap
```

```python
# Chargement des données à partir d'excel , grand capteur
capteur_grand = pd.read_csv(r"C:\Users\Barp9\OneDrive - Universidad Nacional de Colombia\INSA\4ème annee\Semestre 2\PIR\Calidad del aire mapa/capteur_grand.csv", sep=';')
capteur_grand
```

```python
# Chargement des données dans des vecteurs, grand capteur
df = pd.DataFrame(capteur_grand) #Commande pour extraire des données d'Excel
latitude_grand = df['Latitude'] 
longitude_grand = df['Longitude']
particles1_grand = df['Particulates(1.0)[ug/m3]']
particles2_grand = df['Particulates(2.5)[ug/m3]']
particles10_grand = df['Particulates(10.0)[ug/m3]']
particles_NH3_grand = df["NH3[ug/m3]"]
```

```python
# Particulates(1.0)[ug/m3], grand capteur
#insa_map = folium.Map(location=[43.569861717907294, 1.4677892622642483], zoom_start=16) #couleur
insa_map = folium.Map(location=[43.569861717907294, 1.4677892622642483], tiles="Stamen Toner", zoom_start=16) #noir et blanc

for i in range(len(latitude_grand)):
    if particles1_grand[i] < 2:
        color_hexa = "#A5F720"
    elif particles1_grand[i] < 4:
        color_hexa = "#56ED1F"
    elif particles1_grand[i] < 6:
        color_hexa = "#D9E028"
    elif particles1_grand[i] < 8:
        color_hexa = "#F7E420"
    else:
        color_hexa = "#F89957"
    folium.Circle(location=[latitude_grand[i],longitude_grand[i]],color="purple",fill_color=color_hexa,radius=20,weight=0,fill_opacity=0.3).add_to(insa_map)


insa_map
# Particulates(1.0)[ug/m3], grand capteur
```

```python
# Particulates(2.5)[ug/m3], grand capteur
#insa_map = folium.Map(location=[43.569861717907294, 1.4677892622642483], zoom_start=16) #couleur
insa_map = folium.Map(location=[43.569861717907294, 1.4677892622642483], tiles="Stamen Toner", zoom_start=16)  #noir et blanc

for i in range(len(latitude_grand)):
    if particles2_grand[i] < 2:
        color_hexa = "#A5F720"
    elif particles2_grand[i] < 4:
        color_hexa = "#56ED1F"
    elif particles2_grand[i] < 6:
        color_hexa = "#D9E028"
    elif particles2_grand[i] < 8:
        color_hexa = "#F7E420"
    else:
        color_hexa = "#F89957"
    folium.Circle(location=[latitude_grand[i],longitude_grand[i]],color="purple",fill_color=color_hexa,radius=20,weight=0,fill_opacity=0.3).add_to(insa_map)


insa_map
# Particulates(2.5)[ug/m3], grand capteur
```

```python
# Particulates(10.0)[ug/m3], grand capteur
#insa_map = folium.Map(location=[43.569861717907294, 1.4677892622642483], zoom_start=16) #couleur
insa_map = folium.Map(location=[43.569861717907294, 1.4677892622642483], tiles="Stamen Toner", zoom_start=16) #noir et blanc

for i in range(len(latitude_grand)):
    if particles10_grand[i] < 2:
        color_hexa = "#A5F720"
    elif particles10_grand[i] < 4:
        color_hexa = "#56ED1F"
    elif particles10_grand[i] < 6:
        color_hexa = "#D9E028"
    elif particles10_grand[i] < 8:
        color_hexa = "#F7E420"
    else:
        color_hexa = "#F89957"
    folium.Circle(location=[latitude_grand[i],longitude_grand[i]],color="purple",fill_color=color_hexa,radius=20,weight=0,fill_opacity=0.3).add_to(insa_map)


insa_map
# Particulates(10.0)[ug/m3], grand capteur
```

```python
# Particulates(NH3)[ug/m3], grand capteur
#insa_map = folium.Map(location=[43.569861717907294, 1.4677892622642483], zoom_start=16) #couleur
insa_map = folium.Map(location=[43.569861717907294, 1.4677892622642483], tiles="Stamen Toner", zoom_start=16)  #noir et blanc

for i in range(len(latitude_grand)):
    if particles_NH3_grand[i] < 200:
        color_hexa = "#A5F720"
    elif particles_NH3_grand[i] < 400:
        color_hexa = "#56ED1F"
    elif particles_NH3_grand[i] < 600:
        color_hexa = "#D9E028"
    elif particles_NH3_grand[i] < 800:
        color_hexa = "#F7E420"
    else:
        color_hexa = "#F89957"
    folium.Circle(location=[latitude_grand[i],longitude_grand[i]],color="purple",fill_color=color_hexa,radius=20,weight=0,fill_opacity=0.3).add_to(insa_map)

insa_map
# Particulates(NH3)[ug/m3], grand capteur
```
