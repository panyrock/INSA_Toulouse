```python
# Librairies
import folium  #Cercles colorés
from folium.features import DivIcon
import sys
import numpy
import pandas as pd #Excel
import numpy as np
from folium.plugins import MiniMap
import leafmap.foliumap as leafmap
```

```python
# Chargement des données à partir d'excel , grand capteur
data_water = pd.read_csv(r"C:\Users\Barp9\OneDrive - Universidad Nacional de Colombia\INSA\4ème annee\Semestre 2\PIR\Calidad del agua mapa/water_data_map.csv", sep=';')
data_water
```

```python
# Chargement des données dans des vecteurs, grand capteur
df = pd.DataFrame(data_water) #Commande pour extraire des données d'Excel
latitude = df['Latitude'] 
longitude = df['Longitude']
Nitrates = df['Amount of nitrates (mg/L [ppm])']
```

```python
# nitrates
#insa_map = folium.Map(location=[43.6043, 1.4437], zoom_start=13) #couleur
insa_map = folium.Map(location=[43.6043, 1.4437], tiles="Stamen Toner", zoom_start=13)  #noir et blanc

#Ultimo punto
#folium.Circle(location=[latitude[9],longitude[9]],color="black",fill_color='#01E13C',radius=80,weight=0.5,fill_opacity=0.9).add_to(insa_map)

for i in range(len(latitude)):
    if Nitrates[i] == 0.5:
        color_hexa = "#01E13C"        
    elif Nitrates[i] == 1:
        color_hexa = "#FAEE00"
    elif Nitrates[i] == 2:
        color_hexa = "#FB870D"        
    elif Nitrates[i] < 5:
        color_hexa = "#F1418C"
    else:
        color_hexa = "#E30D7C"
    folium.Circle(location=[latitude[i],longitude[i]],color="black",fill_color=color_hexa,radius=150,weight=0.5,fill_opacity=0.9).add_to(insa_map)


#___________________________________________

# Definir los elementos y sus colores
legend_dict = {
    'Amount of nitrates 0.5 mg/L': '#01E13C',
    'Amount of nitrates 1 mg/L': '#FAEE00',
    'Amount of nitrates 2 mg/L': '#FB870D'
}

# Crear la leyenda
legend_html = '<div style="position:fixed; bottom:200px; left:500px; z-index:9999; font-size:25px; background-color:white; padding: 5px 10px; border-radius:10px;">'
legend_html += '<h1>Nitrates, earthwatch tests</h1>'  # Aquí agregamos el título
for elem, color in legend_dict.items():
    legend_html += f'<div style="display:inline-block; background-color:{color}; width:35px; height:35px; margin-right:5px;"></div><div style="display:inline-block;">{elem}</div><br>'

legend_html += '</div>'

# Agregar la leyenda al mapa
insa_map.get_root().html.add_child(folium.Element(legend_html))

# Mostrar el mapa
 
    
insa_map
# Nitrates
```
